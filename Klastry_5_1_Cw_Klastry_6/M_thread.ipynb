```python
# Ćwiczenie 1: Współzawodnictwo o licznik - przykład

import threading

counter = 0

def incrementer():
    global counter
    for _ in range(100_000):
        counter += 1

def decrementer():
    global counter
    for _ in range(100_000):
        counter -= 1

t1 = threading.Thread(target=incrementer)
t2 = threading.Thread(target=decrementer)

t1.start()
t2.start()

t1.join()
t2.join()

print(f"Końcowa wartość licznika: {counter}")
```

```python
# Ćwiczenie 1: Współzawodnictwo o licznik - mod 'time.sleep(0)' - częstszy błąd

import threading
import time

counter = 0

def incrementer():
    global counter
    for _ in range(100_000):
        counter += 1
        time.sleep(0)

def decrementer():
    global counter
    for _ in range(100_000):
        counter -= 1
        time.sleep(0)

t1 = threading.Thread(target=incrementer)
t2 = threading.Thread(target=decrementer)

t1.start()
t2.start()

t1.join()
t2.join()

print(f"Końcowa wartość licznika: {counter}")
```

```python
# Ćwiczenie 1: Współzawodnictwo o licznik - mod 'Lock'

import threading

counter = 0
lock = threading.Lock()

def incrementer():
    global counter
    for _ in range(100_000):
        with lock:
            counter += 1

def decrementer():
    global counter
    for _ in range(100_000):
        with lock:
            counter -= 1

t1 = threading.Thread(target=incrementer)
t2 = threading.Thread(target=decrementer)

t1.start()
t2.start()

t1.join()
t2.join()

print(f"Końcowa wartość licznika: {counter}")
```

```python
# Ćwiczenie 2: Współdzielona lista - przykład

import threading
import time
tasks = [f"Zadanie {i}" for i in range(10)]
def worker(name):
    while tasks:
        try:
 # Pobierz pierwsze zadanie z listy
            task = tasks.pop(0)
            print(f"Wątek {name} pobiera: {task}")
 # Symuluj pracę nad zadaniem
            time.sleep(0.01)
        except IndexError:
 # Ta sytuacja nie powinna mieć miejsca, ale...
            print(f"Wątek {name} próbował pobrać zadanie z pustej listy!")
            break
t1 = threading.Thread(target=worker, args=("A",))
t2 = threading.Thread(target=worker, args=("B",))

t1.start()
t2.start()

t1.join()
t2.join()

print("Wszystkie zadania ukończone.")
```

```python
# Ćwiczenie 2: Współdzielona lista - mod

import threading
import time

tasks = [f"Zadanie {i}" for i in range(10)]
lock = threading.Lock()

def worker(name):
    while True:
        with lock:
            if not tasks:
                break
            task = tasks.pop(0)

        print(f"Wątek {name} pobiera: {task}")

        # Symulacja pracy
        time.sleep(0.01)

t1 = threading.Thread(target=worker, args=("A",))
t2 = threading.Thread(target=worker, args=("B",))

t1.start()
t2.start()

t1.join()
t2.join()

print("Wszystkie zadania ukończone.")
```

```python
# Ćwiczenie 3: Zakleszczenie (Deadlock) - przykład

import threading
import time

lock_a = threading.Lock()
lock_b = threading.Lock()

def thread_1():
    print("Wątek 1: Próbuję zdobyć blokadę A...")
    with lock_a:
        print("Wątek 1: Zdobyłem blokadę A. Czekam 1s...")
        time.sleep(1)
        print("Wątek 1: Próbuję zdobyć blokadę B...")
        with lock_b:
            print("Wątek 1: Zdobyłem obie blokady!")

def thread_2():
    print("Wątek 2: Próbuję zdobyć blokadę B...")
    with lock_b:
        print("Wątek 2: Zdobyłem blokadę B. Czekam 1s...")
        time.sleep(1)
        print("Wątek 2: Próbuję zdobyć blokadę A...")
        with lock_a:
            print("Wątek 2: Zdobyłem obie blokady!")

t1 = threading.Thread(target=thread_1)
t2 = threading.Thread(target=thread_2)

t1.start()
t2.start()

t1.join()
t2.join()

print("Program zakończony pomyślnie.")
```
